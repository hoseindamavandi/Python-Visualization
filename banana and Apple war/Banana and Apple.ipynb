```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.rcParams['xtick.labelsize'] = 15
plt.rcParams['ytick.labelsize'] = 15
plt.rcParams['legend.fontsize'] = 15
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['axes.titlesize'] = 15
```

```python
N = 1000
std = 0.6
```

```python
apple_width = np.random.normal(6, std, N)
apple_length = np.random.normal(6, std, N)

banana_width = np.random.normal(4, std, N)
banana_length = np.random.normal(8, std, N)

width = np.concatenate((apple_width, banana_width))
length = np.concatenate((apple_length, banana_length))

X = np.array([length,width]).T
apple_label = np.zeros(N, dtype='int')
banana_label = np.ones(N, dtype='int')

Y = np.concatenate((apple_label, banana_label))
```

```python
plt.xlabel('Length')
plt.ylabel('Width')
plt.scatter(X[:N, 0],X[:N,1], c = 'orange', marker='o')
plt.scatter(X[N:, 0],X[N:,1], c = 'blue', marker='^')
```

```python
labels = ['bananas','apples','bananas','apples']
x = [1, 1.5, 2.5, 3]
height = [banana_length.mean() ,apple_length.mean(), banana_width.mean(), apple_width.mean() ]
yerr = [banana_length.std() ,apple_length.std() , banana_width.std(), apple_width.std()]
color = ['tab:blue','tab:blue','tab:orange','tab:orange']
plt.bar(x=x,
        height=height,
        color=color,
        width=0.4,
        error_kw={'lw': 4},
        yerr=yerr)

plt.xticks(ticks=x, labels=labels)
plt.title('Fruit length VS width')
plt.grid()
```

```python
plt.boxplot(x=banana_length, patch_artist=True, boxprops={'facecolor': 'tab:blue'}, positions=[0], labels=['bananas'], widths=[0.2])
plt.boxplot(x=apple_length, patch_artist=True, boxprops={'facecolor': 'tab:orange'}, positions=[0.3], labels=['apples'], widths=[0.2])
plt.ylabel('length')
plt.xlim(-0.2, 0.5)
plt.title('Length Distribution ')
```

```python
plt.hist(x=banana_length, bins=30, density=True, color='tab:blue', label='bananas')
plt.hist(x=apple_length, bins=30, density=True, color='tab:orange', label='apples', alpha=0.8)
plt.legend(frameon=False)
plt.title('Width Distribution ')
plt.xlabel('Width')
```

```python
fig, (ax, cax) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [0.9, 0.1]})
_, _, _, mappable = ax.hist2d(
    x=length, y=width, bins=50, cmap=plt.cm.hot)
ax.set_xlabel('length')
ax.set_ylabel('width')
plt.colorbar(mappable=mappable, cax=cax, ax=ax)
```

```python
b= banana_length/banana_width
a = apple_length/apple_width
plt.hist(x=b,
         bins=50, density=True, color='tab:blue', label='banana')
plt.hist(x=a,
         bins=50, density=True, color='tab:orange', label='apple', alpha=0.8)
plt.legend()
plt.title('L/W Distribution ')
plt.xlabel('L/W ratio')
```

```python
b=banana_width*banana_length
a=apple_width*apple_length
plt.hist(x=b, histtype='step', lw=4,
         bins=40, density=True, color='tab:blue', label='bananas')
plt.hist(x=a, histtype='step', lw=4,
         bins=40, density=True, color='tab:orange', label='apples')
plt.title('Area Distribution ')
plt.legend()
plt.xlabel('Area')
```

```python
b = 2*(banana_length+banana_width)
a = 2*(apple_length+apple_width)

plt.hist(x=b, histtype='step', lw=4,
         bins=50, density=True, color='tab:blue', label='bananas')
plt.hist(x=a, histtype='step', lw=4,
         bins=50, density=True, color='tab:orange', label='apples')
plt.title('circumference Distribution ')
plt.legend()
plt.xlabel('Circumference')
```
